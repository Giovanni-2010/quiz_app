```python
import pandas as pd
```

```python
df = pd.read_json('questions.json',encoding='utf-8')
```

```python
df.duplicated(subset='question').sum()
```

```python
df.drop_duplicates(subset=['question'],inplace=True)
df.shape
```

```python
total_questions = len(df)
unique_topics = df['topic'].nunique()
questions_per_topic = df['topic'].value_counts()
questions_per_type = df['type'].value_counts()
all_references = df['reference'].unique()
```

```python
print(f"Total Questions: {total_questions}")
print(f"Unique Topics: {unique_topics}")
print("\nQuestions per Topic:")
print(questions_per_topic)
print("\nQuestions per Type:")
print(questions_per_type)
print("\nReferences:")
for ref in all_references:
    print(f" - {ref}")
```

```python
df.info()
```

```python
df = df.replace({'True': 'صح', 'False': 'خطأ'})
```

```python
def translate_list(row):
    if row['type'] == 'boolean':
        row['choices'] = ['صح', 'خطأ']
    return row
```

```python
df = df.apply(translate_list, axis=1)
```

```python
df.sample(10)
```

```python
df.shape
```

```python
df['id'] = range(1, 395)
```

```python
df['question'] = df['question'].str.replace(r'\؟$', '', regex=True)
df.head()
```

```python
df.to_json('questions_final.json', orient='records', force_ascii=False, indent=2)
```
